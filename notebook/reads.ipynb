# Realizando Consultas (Equiv Select)

```python
from connect import db
from bson.objectid import ObjectId

libros = db.libros
```

```python
# Encontrar todos las colecciones
for doc in libros.find():
    print(doc)
```

```python
# Filtro con where

for doc in libros.find({'autor': 'Mario Vargas Llosa'}):
    print(doc['titulo'], doc['anio'])
```

```python
# Comparar
# Publicaciones mayores a√±o 1960
for doc in libros.find({'anio': {'$gte':1960}}):
    print(doc['titulo'], doc['anio'])
```
