# Insertar Datos
- InsertOne
- InserMany

```python
# Importando conexion
from connect import db

libros = db.libros

# Estructura coleccion (Equiv. Estructura Tabla)
libro = {
    'titulo': 'La casa Verde',
    'autor': 'Mario Vargas Llosa',
    'anio': 1965,
    'categorias': [
        'novela', 'ficcion' ] ,
    'stock': 4
}

# Insert One
res = libros.insert_one(libro)
print('InsertOne id:', res.inserted_id)
```

```python
# Insert Many (Varias inserciones)
docs = [
    #Libro 2
    {
    'titulo': 'don Quijote',
    'autor': 'Miguel de Cervantes Saavedra',
    'anio': 1605,
    'categorias': [
        'clásico' ] ,
    'stock': 5
    },
    # Libro 3
    {
    'titulo': 'La ciudad y los perros',
    'autor': 'Mario Vargas Llosa',
    'anio': 1963,
    'categorias': [
        'novela'] ,
    'stock': 3
    }
]

res_many= libros.insert_many(docs)
print('InserMany ids: ', res_many.inserted_ids)
```
